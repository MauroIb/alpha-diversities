```python
import glob, sys, os
import pandas as pd
from pathlib import Path

### Funcion que combina fastamanifest con runtable creando metadata valido.

archivo_manifest = 'qiime/fasta_manifest.csv'
archivo_metadata = 'SraRunTable.txt'

### Nombre de columna que contiene nombres fastq
label_metadata = 'Run'
label_manifest = 'absolute-filepath'

### Cargo archivos
meta=pd.read_csv(archivo_metadata,sep ='\t')
#display(meta.head())
manifest=pd.read_csv(archivo_manifest)
#display(manifest.head())

### Recorto nombre archivos
manifest[label_manifest] = [Path(x).stem.split('.')[0] for x in manifest[label_manifest]]
#display(manifest.head())

### Si es paired end, borrar columna con direccion y sufijos _1/2
if 'direction' in manifest:
    manifest[label_manifest] = [x.split('_')[0] for x in manifest[label_manifest]]
    manifest.drop_duplicates(subset=label_manifest, inplace=True)
    manifest.drop(columns='direction', inplace=True)
    #display(manifest.head())
    
### Combino con sample frequency


### Combino y exporto
result = pd.merge(manifest, meta, left_on=label_manifest, right_on=label_metadata)
result.drop(columns=label_manifest, inplace=True)
result.to_csv('sample-metadata.tsv',sep='\t', index=False)
display(result)
```
